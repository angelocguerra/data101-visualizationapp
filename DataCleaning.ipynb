## Data Cleaning

- Remove the columns associated with specified sexes since the focus of the data is not to drill-down on those
- Rename the columns to only indicate the year
- Remove the 'Philippines' row since the visualization targets drilling down data at a regional level

```python
# Remove unrelated columns
primary_completion_cleaned = primary_completion.drop(['Boys 2006','Boys 2007','Boys 2008','Boys 2009','Boys 2010','Boys 2011','Boys 2012','Boys 2013','Boys 2014','Boys 2015','Girls 2006','Girls 2007','Girls 2008','Girls 2009','Girls 2010','Girls 2011','Girls 2012','Girls 2013','Girls 2014','Girls 2015'], axis=1)

# Rename columns
primary_completion_cleaned = primary_completion_cleaned.rename(columns={"Both Sexes 2006": "2006", "Both Sexes 2007": "2007", "Both Sexes 2008": "2008", "Both Sexes 2009":"2009", "Both Sexes 2010":"2010", "Both Sexes 2011":"2011", "Both Sexes 2012":"2012", "Both Sexes 2013":"2013", "Both Sexes 2014":"2014", "Both Sexes 2015":"2015"})

# Drop Philippines row
primary_completion_cleaned = primary_completion_cleaned.query("Region != 'Philippines'")
primary_completion_cleaned
```

```python
# Save as new CSV
primary_completion_cleaned.to_csv('cleaned-datasets/Primary/Primary_Completion_Rate_by_Region_and_Year.csv', sep=',', encoding='utf-8')
```
