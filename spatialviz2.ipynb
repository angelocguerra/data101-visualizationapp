```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd

import plotly.graph_objects as go
from plotly.subplots import make_subplots

import shapely

from pathlib import Path
import plotly.express as px
%pylab inline
```

```python
dataset_folder = Path('cleaned-datasets/')
```

```python
primary_completion = gpd.read_file(dataset_folder / 'Spatial/gpd/primary_completion.shp')
```

```python
primary_completion = primary_completion.to_crs('EPSG:4326')
primary_completion = primary_completion[primary_completion.is_valid]
primary_completion['geometry'] = primary_completion['geometry'].buffer(0)
```

```python
fig = px.choropleth(primary_completion, 
                    geojson=primary_completion.geometry,
                    center={'lat': 12.099568, 'lon': 122.733168}, 
                    locations=primary_completion.index,
                    color='2006')
```

```python
fig.show()
```
